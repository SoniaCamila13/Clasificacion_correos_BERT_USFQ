Se carga el csv del dataset para anonimizar los datos:

```python
import pandas as pd

df = pd.read_csv("bases_unificadas.csv", sep=';')
df.head()
```

```python
df.shape
```

Se anonimizan direcciones de correo, cédulas, teléfonos y nombres

```python
import re

def anonymize_text(text):

    if pd.isna(text):
        return ""

    text = str(text)

    # emails
    text = re.sub(r'\S+@\S+', '[EMAIL]', text)

    # cédulas o números largos
    text = re.sub(r'\b\d{7,13}\b', '[ID]', text)

    # teléfonos
    text = re.sub(r'\b\d{9,10}\b', '[PHONE]', text)

    # nombres después de frases comunes
    text = re.sub(r'(mi nombre es\s+)([A-ZÁÉÍÓÚÑ][a-záéíóúñ]+\s[A-ZÁÉÍÓÚÑ][a-záéíóúñ]+)', r'\1[NOMBRE]', text, flags=re.IGNORECASE)

    text = re.sub(r'(soy\s+)([A-ZÁÉÍÓÚÑ][a-záéíóúñ]+\s[A-ZÁÉÍÓÚÑ][a-záéíóúñ]+)', r'\1[NOMBRE]', text, flags=re.IGNORECASE)

    return text
```

```python
df["Asunto"] = df["Asunto"].apply(anonymize_text)
df["Cuerpo"] = df["Cuerpo"].apply(anonymize_text)
df["Remitente"] = df["Remitente"].apply(anonymize_text)
```

```python
df.head()
```

```python
df.to_csv("correos_anonimizados.csv", index=False)
```
